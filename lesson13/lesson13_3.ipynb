```python
import numpy as np
import pandas as pd
np.random.seed(123)
scores = np.random.randint(50,101,size=(50,5))
students = pd.DataFrame(scores,
            index=[f'學生{num}' for num in range(1,51)],
            columns = ['國文','英文','數學',"生物",'物理'])
students
```

```python
students["國文"] #subscribes
students[['國文','數學']] #subscribes裡面加list,提取出不只一個值
```

```python
sum_series = np.sum(students,axis=1)
average_series = np.mean(students, axis=1)
average_series
```

```python
students_finish = students.copy()
students_finish['總分'] = sum_series
students_finish['平均'] = average_series
students_finish
```

```python
#由index開始選取
students_finish.loc['學生1'] #利用索引，抓取單筆資料
students_finish.loc[["學生1",'學生3']] #利用索引
students_finish.loc['學生3':'學生10'] #利用slicing(切割)抓取某一串連續的資料
students_finish.loc['學生1':'學生10','國文':'物理'] #利用slicing(切割)抓取某一串連續的資料 (欄和列)
students_finish.iloc[2:10,:5] #利用iloc 抓取資料; 冒號前面如果沒有打數字就預設重第一筆開始
```

```python
students = students_finish.loc['學生1'] #利用索引
students['國文']=100 # 以下提示是在說明，用loc 或是 iloc 抓取出來的資料都是"View"(想像成唯讀)，修改國文成為100不會被回寫到原本的資料裡面
students_finish #原本的資料students_finish的學生1國文還是95
```

```python
#用loc 或 iloc 抓取出的資料如果要修改，要先用.copy將抓取出來的資料轉成"實體"
students10 = students_finish.iloc[:10].copy()
students10.loc['學生1','國文'] = 100
students10
```

```python
#透過Mask選取資料
#直接給Mask bool值,選出國文84分以上的
mask = [True,False,False,True,True,False,False,False,True,True]
students10[mask]
```

```python
#透過Mask選取資料
#給Mask資料的定義,選出國文84分以上的
mask = students10['國文'] >= 85
students10[mask]
```

```python
students10[students10['數學'] >= 90]
```

```python
students_finish[students_finish['總分'] >= 400]
```
