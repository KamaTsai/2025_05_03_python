```python
import matplotlib.pyplot as plt

labels= ['Nokia', 'Samaung', 'Apple','Lumia']
values= [20,30,45,10]
colors=['gray','green','red','black']
explode =(0,0,0.1,0)

_,_, autotexts= plt.pie(values, labels=labels, colors=colors, explode=explode, shadow=True,startangle=180, autopct='%1.1f%%' )
plt.setp(autotexts, size=12, color="white")
plt.show()
```
